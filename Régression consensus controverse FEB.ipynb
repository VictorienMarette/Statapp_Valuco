```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')

df2['period'] = df2['period'].astype(str).str[:7]
df3['period'] = '2025-02'

df2['isin'] = df2['isin'].str.strip()
df3['ISSUER_ISIN'] = df3['ISSUER_ISIN'].str.strip()

df2_fevrier = df2[(df2['period'] == '2025-02') & (df2['score_kind'] == 'ESG')]
df2_fevrier = df2_fevrier.drop_duplicates(subset='isin')

df = pd.merge(df3, df2_fevrier, left_on='ISSUER_ISIN', right_on='isin')

X = df[[
    'CONTR_SUPPLY_CHAIN_LABOR_N_TOTAL',
    'CONTR_TOXIC_EMISS_WASTE_N_TOTAL',
    'CONTR_WATER_N_TOTAL',
    'TAX_TRANSP_CONTROVERSIES_TOTAL'
]]
y = df['consensus']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = LinearRegression()
model.fit(X_scaled, y)

print(model.coef_)
print(model.intercept_)
print(model.score(X_scaled, y))

X_vif = pd.DataFrame(X_scaled, columns=X.columns)
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print("\nVIF :\n", vif_data)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')

df2['period'] = df2['period'].astype(str).str[:7]
df3['period'] = '2025-02'

df2['isin'] = df2['isin'].str.strip()
df3['ISSUER_ISIN'] = df3['ISSUER_ISIN'].str.strip()

df2_fevrier = df2[(df2['period'] == '2025-02') & (df2['score_kind'] == 'ESG')]
df2_fevrier = df2_fevrier.drop_duplicates(subset='isin')

df = pd.merge(df3, df2_fevrier, left_on='ISSUER_ISIN', right_on='isin')

df = df.dropna(subset=['CONTR_SUPPLY_CHAIN_LABOR_N_TOTAL', 'CONTR_TOXIC_EMISS_WASTE_N_TOTAL', 'CONTR_WATER_N_TOTAL', 'TAX_TRANSP_CONTROVERSIES_TOTAL', 'consensus'])

X = df[[
    'CONTR_SUPPLY_CHAIN_LABOR_N_TOTAL',
    'CONTR_TOXIC_EMISS_WASTE_N_TOTAL',
    'CONTR_WATER_N_TOTAL',
    'TAX_TRANSP_CONTROVERSIES_TOTAL'
]]
y = df['consensus']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = LinearRegression()
model.fit(X_scaled, y)

print(model.coef_)
print(model.intercept_)
print(model.score(X_scaled, y))

X_vif = pd.DataFrame(X_scaled, columns=X.columns)
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print("\nVIF :\n", vif_data)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')

df2_bunzl = df2[df2['name'] == 'BUNZL PUBLIC LIMITED COMPANY']
df3_bunzl = df3[df3['ISSUER_NAME'] == 'BUNZL PUBLIC LIMITED COMPANY']

df = pd.merge(df3_bunzl, df2_bunzl, left_on='ISSUER_ISIN', right_on='isin')

X = df[['CONTR_SUPPLY_CHAIN_LABOR_N_TOTAL', 'CONTR_TOXIC_EMISS_WASTE_N_TOTAL', 'TAX_TRANSP_CONTROVERSIES_TOTAL']]
y = df['consensus']

X = X.fillna(0)
y = y.astype(float)

model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
print(model.score(X, y))

for col in X.columns:
    sns.regplot(x=X[col], y=y)
    plt.title(col + ' vs consensus')
    plt.show()
```
