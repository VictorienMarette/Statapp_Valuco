```python
import pandas as pd
import matplotlib.pyplot as plt

def stat_desc(type_score):

    # On charge les données
    df= pd.read_csv('ValueCo data extract - ENSAE research project.csv')

    # On filtre par 'score_kind'
    df_filtré = df[df['score_kind'] == type_score]

    # On calcule les statistiques descriptives
    stats_descriptives = df.describe()
    print(stats_descriptives)

    # On trace un Histogramme
    df_filtré['consensus'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
    plt.title('Histogramme de Consensus')
    plt.xlabel('Consensus')
    plt.ylabel('Fréquence')
    plt.show()

    # On trace un Boxplot
    df_filtré.boxplot(column='consensus', patch_artist=True, notch=True, vert=False)
    plt.title('Boxplot de Consensus')
    plt.show()

    # On utilise Scatter pour observer la relation entre période et consensus 
    df_filtré.plot(kind='scatter', x='period', y='consensus', color='purple')
    plt.title('Relation entre Période et Consensus')
    plt.xticks(rotation=90)
    plt.xlabel('Période')
    plt.ylabel('Consensus')
    plt.show()

    # On trace la courbe de Densité
    df_filtré['consensus'].plot(kind='kde', color='green')
    plt.title('Courbe de Densité de Consensus')
    plt.xlabel('Consensus')
    plt.show()
```

```python
stat_desc('E')
```

```python
stat_desc('S')
```
