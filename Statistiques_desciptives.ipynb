```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


def stat_desc(type_score):

    # On charge les données
    df= pd.read_csv('BQ Results Apr 8 2025.csv')

    # On filtre par 'score_kind'
    df_filtré = df[df['score_kind'] == type_score]

    # On calcule les statistiques descriptives
    stats_descriptives = df_filtré.describe()
    print(stats_descriptives)

    # Créer l'histogramme
    plt.hist(df_filtré['consensus'], bins=30, density=True, alpha=0.6, color='g')

    # Ajouter la courbe de densité
    sns.kdeplot(df_filtré['consensus'], color='blue', linewidth=2)
    plt.title("Histogramme et courbe de densité du consensus du score")
    plt.xlabel("Consensus")
    plt.ylabel("Densité")
    plt.savefig(f'./Graphiques/Histogramme et courbe de densité du consensus du score {type_score} de toutes les entreprises.png')
    plt.show()

    # On trace un Boxplot
    df_filtré.boxplot(column='consensus', patch_artist=True, notch=True, vert=False)
    plt.title('Boxplot de Consensus')
    plt.savefig(f'./Graphiques/Boxplot du consensus des score {type_score} de toutes les entreprises.png')
    plt.show()

    # On utilise Scatter pour observer la relation entre période et consensus 
    df_filtré.plot(kind='scatter', x='period', y='consensus', color='purple')
    plt.title('Relation entre Période et Consensus')
    plt.xticks(rotation=90)
    plt.xlabel('Période')
    plt.ylabel('Consensus')
    plt.savefig(f'./Graphiques/Consensus des scores {type_score} de toutes les entreprises au cours du temps.png')
    plt.show()
```

```python
stat_desc('E')
```
