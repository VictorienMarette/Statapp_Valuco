```python
import pandas as pd
import matplotlib.pyplot as plt

# On charge les donnes
df = pd.read_csv('ValueCo data extract - ENSAE research project.csv')

# On calcule les statistiques descriptives
stats_descriptives = df.describe()
print(stats_descriptives)

# On trace un Histogramme
df['consensus'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Histogramme de Consensus')
plt.xlabel('Consensus')
plt.ylabel('Frquence')
plt.show()

# On trace un Boxplot
df.boxplot(column='consensus', patch_artist=True, notch=True, vert=False)
plt.title('Boxplot de Consensus')
plt.show()

# On utilise Scatter pour observer la relation entre priode et consensus 
df.plot(kind='scatter', x='period', y='consensus', color='purple')
plt.title('Relation entre Priode et Consensus')
plt.xticks(rotation=90)
plt.xlabel('Priode')
plt.ylabel('Consensus')
plt.show()

# On trace la courbe de Densit
df['consensus'].plot(kind='kde', color='green')
plt.title('Courbe de Densit de Consensus')
plt.xlabel('Consensus')
plt.show()
```
